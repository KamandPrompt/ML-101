# Regression Problem

        Reading the file

```python
import pandas as pd
import numpy as np
data=pd.read_csv("ML101_train_dataset.csv")
data.corr()
data.dropna(subset=["LifeStyle"],inplace=True)
data.dropna(subset=["Systolic BP","Diastolic BP"],inplace=True)
data.drop(columns=["Average Daily Steps","Hours of Sleep"],inplace=True)
a=7/3
data.dropna(thresh=a, inplace=True)
data
```

        Filling out missing values in Categorical data

```python
lp=list(data['LifeStyle'])
#Dropping the lifestyle column
d2=data.drop("LifeStyle",axis=1)
```

```python
#code to put mode in missing values of gender attribute
mode = d2['Gender'].mode()[0]
d2['Gender'].fillna(mode, inplace=True)

#Dropping the gender column
d3=d2.drop("Gender",axis=1)
```

        Interpolating numerical data

```python
d3.interpolate(method='linear',axis=0,inplace=True)
mat=d3.to_numpy()
len(mat[0])
```

        Removing Outliers

```python
import matplotlib.pyplot as plt
def remove_outlier(matrix):
    c=0
    for i in range(len(matrix[0])):  # Iterate over columns
        l1 = list(matrix[:, i])
        # plt.hist(l1)
        # plt.show()          # only first two att not normal others are normal
        me=np.mean(l1)
        q1 = np.percentile(l1, 25)
        q2 = np.percentile(l1, 50)
        q3 = np.percentile(l1, 75)
        iqr = q3 - q1
        for j in range(len(l1)):
            if (l1[j] > q3 + 1.5 * iqr or l1[j] < q1 - 1.5 * iqr):  # outlier if val>q3=1.5*iqr or val<q1-1.5*iqr
                if i==0 or i==1:
                    l1[j] = q2  # Replaced with median
                else:
                    l1[j]=me #Replace with mean
                c+=1
        matrix[:, i] = l1
    print(c)
    return matrix
```

```python
d5=pd.DataFrame(remove_outlier(d3.to_numpy()[:,:6]),columns=d3.columns[:6])
remove_outlier(d5.to_numpy())
d5["Systolic BP"]=list(d3["Systolic BP"])
d5["Diastolic BP"]=list(d3["Diastolic BP"])
```

        Adding the modified gender column back to data

```python
d5["Gender"]=list(d2["Gender"])           #Adding back the modified column of gender
d5
```

        Converting categorical data into numerical by using One hot Encoding

```python
one_hot_encoded = pd.get_dummies(d5, columns=["Gender"])
one_hot_encoded
```

        Splitting into test and train data 

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(one_hot_encoded, one_hot_encoded["Gender_Male"],one_hot_encoded["Gender_Female"]):
    strat_train_set = one_hot_encoded.loc[train_index]
    strat_test_set = one_hot_encoded.loc[test_index]
strat_train_set["Gender_Male"].value_counts()  #All ratios are same after splitting into training and test  so splitting done is correct
strat_test_set
```

    Building the regression model

```python
y1=np.array(strat_train_set["Systolic BP"])
y2=np.array(strat_train_set["Diastolic BP"])
red_data=strat_train_set.drop(columns=["Systolic BP","Diastolic BP"],axis=1)
red=red_data.to_numpy()
xtx=np.linalg.inv(np.matmul(np.transpose(red),red))
xty1=np.matmul(np.transpose(red),y1)
xty2=np.matmul(np.transpose(red),y2)
w1=np.matmul(xtx,xty1)
w2=np.matmul(xtx,xty2)
```

     Checking the accuracy on test data splitted

```python
y1_true=np.array(strat_test_set["Systolic BP"])
y2_true=np.array(strat_test_set["Diastolic BP"])
red_data=strat_test_set.drop(columns=["Systolic BP","Diastolic BP"],axis=1)
red=red_data.to_numpy()

y1_predict=list(np.matmul(red,w1))
y2_predict=np.matmul(red,w1)
```

```python
from sklearn.metrics import mean_squared_error

# Calculate the mean squared error
mse = mean_squared_error(y1_true, y1_predict)
mse
```

# Classification Problem



        Preprocessing the data(Above Code)


```python
data=pd.read_csv("ML101_train_dataset.csv")
data.dropna(subset=["LifeStyle"],inplace=True)

lp=list(data['LifeStyle'])
#Dropping the lifestyle column
d2=data.drop(["LifeStyle","Systolic BP","Diastolic BP"],axis=1)

#code to put mode in missing values of gender attribute
mode = d2['Gender'].mode()[0]
d2['Gender'].fillna(mode, inplace=True)

#Dropping the gender column
d3=d2.drop("Gender",axis=1)

d3.interpolate(method='linear',axis=0,inplace=True)
mat=d3.to_numpy()
len(mat[0])

d5=pd.DataFrame(remove_outlier(d3.to_numpy()),columns=d3.columns)
remove_outlier(d5.to_numpy())
d5["Gender"]=d2["Gender"] 
d5
```

        converting categorical data of gender into numerical by one hot encoding and removing diastolic and systolic columns

```python
 
one_hot = pd.get_dummies(d5, columns=["Gender"])
one_hot["LifeStyle"]=lp
one_hot
```

        Splitting data into train and test

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(one_hot, one_hot["Gender_Male"],one_hot["Gender_Female"]):
    x_train = one_hot.loc[train_index]
    x_test = one_hot.loc[test_index]
x_train
```

        Dropping lifestyle and converting data to matrix format

```python
cate=list(x_train["LifeStyle"])
x_train.drop(columns=["LifeStyle"],axis=1,inplace=True)
matt=x_train.to_numpy()
len(matt[0,:])
```

        Classifying data into 4 classes average,bad,good,great

```python
class1 = []
class2 = []
class3 = []
class4 = []

col = x_train.columns

for i in range(len(cate)):
    if cate[i] == "Average":
        class1.append(matt[i,:])
    if cate[i] == "Bad":
        class2.append(matt[i,:])

    if cate[i] == "Good":
        class3.append(matt[i,:])

    if cate[i] == "Great":
        class4.append(matt[i,:])
```

        Calculating Prior Probability

```python
#Prior prob
p_c1=len(class1)/len(x_train)
p_c2=len(class2)/len(x_train)
p_c3=len(class3)/len(x_train)
p_c4=len(class4)/len(x_train)
print(p_c1,p_c2,p_c3,p_c4)
class1=np.array(class1)
class2=np.array(class2)
class3=np.array(class3)
class4=np.array(class4)

len(class1[0,:])
```

        Calculating Mean and Covariance of each class

```python
mean_c1=[]
mean_c2=[]
mean_c3=[]
mean_c4=[]
for i in range(10):
    mean_c1.append(np.mean(class1[:,i]))
    mean_c2.append(np.mean(class2[:,i]))
    mean_c3.append(np.mean(class3[:,i]))
    mean_c4.append(np.mean(class4[:,i]))

cov_c1=np.cov(np.transpose(class1))
cov_c2=np.cov(np.transpose(class2))
cov_c3=np.cov(np.transpose(class3))
cov_c4=np.cov(np.transpose(class4))
```

      Defining function for calculating guassian prob for multivariate distribution

```python
def prob_cal_multi(mean,cov,x):
    a=1/(np.linalg.det(cov)*(((2)*(np.pi))**6))
    b=np.dot(np.dot(np.transpose(x-mean),np.linalg.inv(cov)),(x-mean))
    c=a*(np.exp((-1/2)*b))
    return c
```

```python
true=x_test["LifeStyle"]
x_test.drop(columns=["LifeStyle"],axis=1,inplace=True)
mat_form=x_test.to_numpy()
```

        Implementing Bayes Classifier

```python
predicted_class=[]
c=0
for i in range(len(mat_form)):
    x=mat_form[i,:]
    #likelihood
    p_x_c1=prob_cal_multi(mean_c1,cov_c1,x)
    p_x_c2=prob_cal_multi(mean_c2,cov_c2,x)
    p_x_c3=prob_cal_multi(mean_c3,cov_c3,x)
    p_x_c4=prob_cal_multi(mean_c4,cov_c4,x)
    c+=1
    print(c)

    #total prob
    p=(p_c1*p_x_c1)+(p_c2*p_x_c2)+(p_c3*p_x_c3)+(p_c4*p_x_c4)

    p_c1_x=(p_c1*p_x_c1)/p
    p_c2_x=(p_c2*p_x_c2)/p
    p_c3_x=(p_c3*p_x_c3)/p
    p_c4_x=(p_c4*p_x_c4)/p

    dic_prob={"Average":p_c1_x,"Bad":p_c2_x,"Good":p_c3_x,"Great":p_c4_x}
    sorted_dict = {k: v for k, v in sorted(dic_prob.items(), key=lambda item: item[1],reverse=True)}
    predicted_class.append(list(sorted_dict.keys())[0])
predicted_class
```

        Calculating accuracy of test data

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
onf_mat=metrics.confusion_matrix(true, predicted_class)
print(onf_mat)
print(accuracy_score(true, predicted_class))
```
